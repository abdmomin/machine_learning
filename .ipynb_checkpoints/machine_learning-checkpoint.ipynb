```python
def find_abs(combine):
    actual = combine.sort_values('Share', ascending=False).head()
    pred = combine.sort_values('predictions', ascending=False)
    ps = []
    found = 0
    seen = 1
    for index, row in pred.iterrows():
        if row['Player'] in actual['Player'].values:
            found += 1
            ps.append(found/seen)
        seen += 1
    return sum(ps)/len(ps)
```

```python
find_abs(combine)
```

```python
years = list(range(1991, 2022))
aps = []
all_predictions = []

for year in years[5:]:
    train = data[data['Year'] < year]
    test = data[data['Year'] == year]
    
    reg.fit(train[ind_vars], train['Share'])
    pred = reg.predict(test[ind_vars])
    pred = pd.DataFrame(pred, columns=['predictions'], index=test.index)
    combine = pd.concat([test[['Player', 'Share']], pred], axis=1)
    all_predictions.append(combine)
    aps.append(find_abs(combine))
```

```python
def add_rank(combine):
    combine = combine.sort_values('Share', ascending=False)
    combine['Rank'] = list(range(1, combine.shape[0]+1))
    combine = combine.sort_values('predictions', ascending=False)
    combine['pred_rank'] = list(range(1, combine.shape[0]+1))
    combine['diff'] = combine['Rank'] - combine['pred_rank']
    return combine
```

```python
add_rank(all_predictions[1])
```

```python
ranking = add_rank(all_predictions[1]).sort_values('diff', ascending=False)
```

```python
ranking[ranking['Rank'] <= 5]
```

```python
def backtest(data, model, year, predictor):
    aps = []
    all_predictions = []

    for year in years[5:]:
        train = data[data['Year'] < year]
        test = data[data['Year'] == year]

        model.fit(train[predictor], train['Share'])
        pred = model.predict(test[predictor])
        pred = pd.DataFrame(pred, columns=['predictions'], index=test.index)
        combine = pd.concat([test[['Player', 'Share']], pred], axis=1)
        combine = add_rank(combine)
        all_predictions.append(combine)
        aps.append(find_abs(combine))
        
    return sum(aps)/len(aps), aps, pd.concat(all_predictions)
```

```python
mean_aps, aps, all_predictions = backtest(data, reg, years[5:], ind_vars)
```
